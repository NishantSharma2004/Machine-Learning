```python
import matplotlib.pyplot as plt
```

```python
num1 = [1,2,3,4,5]
num2 = [20,40,60,80,100]
```

```python
plt.plot(num1,num2,linestyle='dotted',linewidth='2',color='black',marker='*',markersize=10,markerfacecolor ='red')
plt.show()
```

```python
plt.plot(num1,color='red',marker='*')
plt.plot(num2,color='blue',marker="o")
plt.xlabel("I am x axis")
plt.ylabel("I am y-axis")
plt.title("Jai Shree Ram")
plt.grid()
```

```python
#scatter chart
num3 = [2,4,6,8,10]
num4 = [20,60,50,40,70]
plt.scatter(num3,num4,color="hotpink")
plt.show()
```

```python
num3 = [2,4,6,8,10]
num4 = [20,60,50,40,70]
plt.bar(num3,num4,color="hotpink")
plt.show()
```

```python
plt.barh(num3, num4, color="blue")
```

```python
plt.pie(num1, labels=['a','b','c','d','e'], colors=["red","black","pink","blue","orange"])
plt.legend()
plt.show()
```
